```python
import numpy as np
# https://github.com/inducer/meshpy/blob/master/examples/jw_meshtools.py
import jw_meshtools as mt
import meshpy.triangle as triangle
import numpy.linalg as la
import matplotlib.pyplot as plt
```

```python
length = 40.
np_1=10
np_2=20

domain_ll, domain_ur = [-50,-50],[150,150]
domain_ll1, domain_ur1 = [20,40],[70,85]


p1,v1=mt.RectangleSegments(domain_ll, domain_ur, num_points=np_1)
p2,v2=mt.RectangleSegments(domain_ll1, domain_ur1, num_points=np_2)
p3,v3=mt.AddCurves(p1,v1,p2,v2)

fname = 'allwells_xyrot.txt'
obs_pts = np.loadtxt('./emilio_mesh/'+fname, usecols=[1,2])
p4,v4 = mt.PointSegments(obs_pts)
p,v=mt.AddCurves(p3,v3,p4,v4)
mesh = mt.DoTriMesh(p,v,edge_length=length, min_angle=28)
mesh_points = np.array(mesh.points)
mesh_elements = np.array(mesh.elements)
fig, ax = plt.subplots(nrows=1, ncols=1, )
ax.triplot(mesh_points[:, 0], mesh_points[:, 1], mesh_elements,) 
ax.plot(obs_pts[:,0], obs_pts[:,1], 'o', color='red', ms=1.5)
ax.set_aspect('equal')
plt.savefig('mesh_with_rect.png', dpi=300)
print('ne: {}'.format(mesh_elements.shape[0]))
print('nn: {}'.format(mesh_points.shape[0]))
print(mesh_elements.shape[0] * 11)
np.savetxt('pts_with_rect.txt',mesh_points)
np.savetxt('elms_with_rect.txt',mesh_elements)
```
